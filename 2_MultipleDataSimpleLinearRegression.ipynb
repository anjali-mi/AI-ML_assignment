# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from mpl_toolkits import mplot3d
```

# Import csv

```python
df=pd.read_csv('/content/Salary_multiple.csv')
```

```python
df.head()
```

# Simple Encoding

```python
df.loc[:,'Field'].replace(['Technical','Non-Technical'],[0,1],inplace=True)
```

```python
print(df.head())
```

# Splitting x and y

```python
#split x and y
x=df.iloc[:,:-1].values
print(x)
y=df.iloc[:,2].values
print(y)
```

```python
x=x.reshape(-1,2)
y=y.reshape(-1,1)
```

# Plot Data

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming 'df' is your DataFrame
# ... (your previous code to load and prepare 'df') ...

# Convert the 'Technical'/'Non-Technical' column to numerical representation
# For example, using Label Encoding:
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['X2_encoded'] = encoder.fit_transform(df.iloc[:,1]) # Encode the second column (index 1)

fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection='3d')

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')

# Update to use the encoded column for y-axis values
sequence_containing_x_values = df.iloc[:,0].values # Assuming first column is X1
sequence_containing_y_values = df['X2_encoded'].values # Using encoded column
sequence_containing_z_values = df.iloc[:,2].values # Assuming third column is Y

ax.scatter(sequence_containing_x_values, sequence_containing_y_values, sequence_containing_z_values)
plt.show()
```

# Assignments

<li> Train </li>
<li> Evaluate Model </li>
<li> Plot the regression line and points </li>

```python
#Split dataset into train and test data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3, random_state=42)


print(x_train.shape)
print(x_test.shape)
```

```python
# Import the necessary model
from sklearn.linear_model import LinearRegression

#train the model
# Instantiate the model
model = LinearRegression()

# Fit the model to the data
# At this point array is Numpy  ndarray.we convert it to array
model = model.fit(x_train, y_train)

print(model.coef_)
print(model.intercept_)
```

```python
#predict on testm data
y_pred = model.predict(x_test)
print(y_pred)
print('*************')
print(y_test)
```

```python
#Evaluate
from sklearn.metrics import r2_score

r_squared = r2_score(y_test, y_pred)
print("The R2 error is :", r_squared)
```

```python
#Predict Custom value
custom_value = np.array([[2, 1]])
print(model.predict(custom_value))
```

```python
#Plotting
fig = plt.figure(figsize=(11, 12))
ax = fig.add_subplot(projection='3d')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')

#  Plot the actual test data points
ax.scatter(x_test[:, 0], x_test[:, 1], y_test, color='blue', label='Actual Test Points')

# Plot the predicted data points
ax.scatter(x_test[:, 0], x_test[:, 1], y_pred, color='red', label='Predicted Points', marker='x', s=100)

ax.plot_surface(x_test[:,0], x_test[:, 1], y_pred,  linewidth=0, antialiased=False, alpha=0.5)


plt.show()
```
